# New Data Comparison

Este código imprime los expocodes de cada archivo netcdf de campañas presuntamente nuevas y campañas ya descargadas, para poder comparar

```python
import os
import xarray as xr
import pandas as pd
```

## Campañas ya descargadas

```python
campana = "SR03"
direct_downloads = "./Data/direct_downloads/"
ruta_old = direct_downloads + campana

# Recorremos cada carpeta
for name in os.listdir(ruta_old):
    if not name.endswith(".nc"):
        continue
    # Extraemos ruta completa
    ruta = os.path.join(ruta_old, name)

    # Abrimos el dataset y extraemos expocode, la fecha de inicio y de fin
    ds = xr.open_dataset(ruta)
    expocode = ds.expocode.values[0]
    fechas = pd.to_datetime(ds.time.values)
    star_date = fechas.min()
    end_date = fechas.max()

    print(f"{name}:")
    print(f"-expocode : {expocode}")
    print(f"-Start Date: {star_date}")
    print(f"-End Date: {end_date}")
    ds.close()
```

## Campañas nuevas

```python
direct_downloads_new = "./Data/direct_downloads_nuevos/"
ruta_new = direct_downloads_new + campana

# Recorremos cada carpeta
for name in os.listdir(ruta_new):
    # Extraemos ruta completa
    ruta = os.path.join(ruta_new, name)

    # Abrimos el dataset y extraemos expocode, la fecha de inicio y de fin
    ds = xr.open_dataset(ruta)
    expocode = ds.expocode.values[0]
    star_date = ds.time.values.min()
    end_date = ds.time.values.max()

    print(f"{name}:")
    print(f"-expocode : {expocode}")
    print(f"-Start Date: {star_date}")
    print(f"-End Date: {end_date}")
    ds.close()
```
