### Sección A05 1981

Limpia de datos y creación de .nc de sección A05 en 1981.

Importamos dependencias

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
```

Directorio con los archivos:

```python
path = "./1981_A05_US008812/"
files = os.listdir(path)
```

En el archivo ***ocldb1743080804.3065585.CTD.nc***, la latitud y longitud tienen varios valores, y ninguna variable, pero en el resto, solo se tiene una latitud y una longitud, junto a 46 variables. Es por esto que vamos a comparar la latitud y longitud del resto de archivos, con el anterior mencionado.

```python
bad_index = -4
bad_file = files[bad_index]
bad_ds = xr.load_dataset(path + bad_file)
total_lat = bad_ds.lat.values
total_lon = bad_ds.lon.values
lats = []
lons = []
for file in files:
    if file == bad_file:
        continue
    f_path = path + file
    ds = xr.load_dataset(f_path)
    if ds.lat.values in total_lat and ds.lon.values in total_lon:
        lats.append(float(ds.lat.values))
        lons.append(float(ds.lon.values))
    else:
        print("No")

lats = sorted(lats)[::-1]
lons = sorted(lons)[::-1]

print ("Latitudes de el resto de archivos:")
print(lats)

print("\n Latitudes de ocldb1743080804.3065585.CTD.nc:")
print(sorted(total_lat)[::-1])

print ("\n longitudes de el resto de archivos:")
print(lons)

print("\n Longitudes de ocldb1743080804.3065585.CTD.nc:")
print(sorted(total_lon)[::-1])
```

___

El archivo ***ocldb1743080804.3065585.CTD.nc*** contiene las longitudes y latitudes que el resto de archivos contienen individualmente (una latitud y longitud por archivo), así que vamos a intentar concatenar todos los archivos en el tiempo. Para ello, primero extraemos el archivo que solo tiene latitudes y longitudes.

```python
extracted_bad_file = files.pop(bad_index)
```

Una vez extraído de la lista de archivos, vamos a concatenar el resto de documentos.

```python
datasets = []
for file in files:
    file_path = path + file
    ds = xr.load_dataset(file_path)
    datasets.append(ds)
```

```python
ds = xr.concat(datasets, dim = "time")
```
