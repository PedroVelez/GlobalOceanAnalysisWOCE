# New Data Comparison
***
Imprime y generá una lista con la información del nombre del archivo, fecha de inicio de la campaña, fecha de finalización de la misma, y su expocode, de archivos guardados en una campaña. En una celda se abren los archivos descargados anteriormente y en la última los archivos descargados nuevos. De esta forma es posible ver si hay archivos que puedan estar repetidos.

```python
# Packages for file manipulation
import os

# Package for data manipulation
import xarray as xr
import pandas as pd
```

### Seleccionar secciones y carpetas

```python
# Select the ocean section
campana = "SR03"

# Old data
direct_downloads = "./Data/direct_downloads/"
ruta_old = direct_downloads + campana

# New data
direct_downloads_new = "./Data/direct_downloads_nuevos/"
ruta_new = direct_downloads_new + campana
```

### Campañas ya descargadas

```python
# Iterate each file
for name in os.listdir(ruta_old):
    if not name.endswith(".nc"):
        continue
    # Extract the complete rute
    ruta = os.path.join(ruta_old, name)

    # Open the dataset and extract the wanted information
    ds = xr.open_dataset(ruta)
    expocode = ds.expocode.values[0]
    fechas = pd.to_datetime(ds.time.values)
    star_date = fechas.min()
    end_date = fechas.max()

    print(f"{name}:")
    print(f"-expocode : {expocode}")
    print(f"-Start Date: {star_date}")
    print(f"-End Date: {end_date}")
    print('----------------------------')
    ds.close()
```

### Campañas nuevas

```python
# Iterate each file
for name in os.listdir(ruta_new):
    # Extract the complete rute
    ruta = os.path.join(ruta_new, name)

    # Open the dataset and extract the wanted information
    ds = xr.open_dataset(ruta)
    expocode = ds.expocode.values[0]
    star_date = ds.time.values.min()
    end_date = ds.time.values.max()

    print(f"{name}:")
    print(f"-expocode : {expocode}")
    print(f"-Start Date: {star_date}")
    print(f"-End Date: {end_date}")
    print('----------------------------')
    ds.close()
```
