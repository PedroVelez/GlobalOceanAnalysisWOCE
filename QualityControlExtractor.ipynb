# Quality Control Extractor
***
El código aquí presentado busca indicar que distintos qc tiene cada fichero.
La idea es con un bucle revisar uno a uno y dar un listado completo de todos las secciones por archivo. Este código lee de ./Data/direct/downloads entendiendose que ya están incluidos los datos nuevos

```python
# Import a special usefull function
from ReadOriginalData import vars_coords_interest

# Import packages for data manipulation
import numpy as np
import xarray as xr

# Import packages for files manipulation
import os

# Ignore warnings
import warnings
warnings.filterwarnings(action='ignore', category =FutureWarning)
```

```python
# Rute of the data
carpeta = "./Data/direct_downloads/"

# Iterate for every directory and file
for carpetas in os.listdir(carpeta):
    print(carpetas)
    for archivos in os.listdir(carpeta + carpetas):
        # Print the file name
        print(archivos)
        
        # Extract the complete rute
        ruta = os.path.join(carpeta, carpetas, archivos)
        
        # Open the dataset
        ds = xr.open_dataset(ruta)

        # Extract  variables, coords and quality controls of interest
        vars_, coords, qcs = vars_coords_interest(ds)
        for qc in qcs:
            if "ctd_temperature" in qc:
                print(f'{qc}: {np.unique(ds[qc].values)}')
            elif "ctd_salinity" in qc:
                print(f'{qc}: {np.unique(ds[qc].values)}')
        print('----------------------------')
        ds.close()
```
