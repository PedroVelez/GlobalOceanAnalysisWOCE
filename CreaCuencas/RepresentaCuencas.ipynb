```python
import xarray as xr
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import cartopy.crs as ccrs
import cartopy.feature as cfeature

from shapely import contains_xy
from shapely import Polygon
```

```python
filesA = ['./Cuencas/NAEast.csv',
'./Cuencas/NAWest.csv',
'./Cuencas/NACaribe.csv',
'./Cuencas/SAEast.csv',
'./Cuencas/SAWest.csv',
'./Cuencas/SAEastNamibia.csv',
'./Cuencas/SAWestMalvinas.csv',
'./Cuencas/SouthAfrica.csv',
'./Cuencas/SACentralAntartica.csv',
'./Cuencas/SPEastAntartica.csv']

filesI=['./Cuencas/EastMadagascar.csv',
'./Cuencas/SouthMadagascar.csv',
'./Cuencas/NIWest.csv',
'./Cuencas/NICentral.csv',
'./Cuencas/NICentralEast.csv',
'./Cuencas/NIEast.csv',
'./Cuencas/SouthAfrica.csv',
#'./Cuencas/SACentralAntartica.csv',
'./Cuencas/SouthAustralia.csv',
'./Cuencas/SIEastAntartica.csv']


filesP=['./Cuencas/NPWest.csv',
'./Cuencas/NPEast.csv',
'./Cuencas/NPEastEast.csv',
'./Cuencas/NPCentral.csv',
'./Cuencas/SPCentral.csv',
'./Cuencas/SPWestCentral.csv',
'./Cuencas/SPEastAustralia.csv',
#'./Cuencas/SIEastAntartica.csv',
'./Cuencas/SPWest.csv',
'./Cuencas/SPEastNorth.csv',
'./Cuencas/SPEastChile.csv']
#'./Cuencas/SouthAustralia.csv',
#'./Cuencas/SPEastAntartica.csv']
```

```python
fig = plt.figure(figsize=(15,10))
ax = plt.axes(projection=ccrs.Robinson())
ax.set_global()
ax.coastlines(resolution='110m')
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')

for file in filesA:
    df = pd.read_csv(file)
    poly=Polygon(np.column_stack((df.lon, df.lat)))
    ax.add_geometries(
        [poly],
        crs=ccrs.PlateCarree(),   # coordenadas lon/lat
        facecolor='red',
        edgecolor='black')
   
    ax.text(df.lon.mean(),df.lat.mean(),file[10:-4],transform=ccrs.PlateCarree(),fontsize=12,ha='center',va='center',bbox=dict(facecolor='white', alpha=0.8,edgecolor='none' ))
    

for file in filesI:
    df = pd.read_csv(file)
    #ax.plot(df.lon, df.lat,color='red', transform=ccrs.PlateCarree())
    poly=Polygon(np.column_stack((df.lon, df.lat)))
    ax.add_geometries(
        [poly],
        crs=ccrs.PlateCarree(),   # coordenadas lon/lat
        facecolor='red',
        edgecolor='black')
    ax.text(df.lon.mean(),df.lat.mean(),file[10:-4],transform=ccrs.PlateCarree(),fontsize=12,ha='center',va='center',bbox=dict(facecolor='white', alpha=0.8,edgecolor='none' ))

for file in filesP:
    df = pd.read_csv(file)
    #ax.plot(df.lon, df.lat,color='red', transform=ccrs.PlateCarree())
    poly=Polygon(np.column_stack((df.lon, df.lat)))
    ax.add_geometries(
        [poly],
        crs=ccrs.PlateCarree(),   # coordenadas lon/lat
        facecolor='red',
        edgecolor='black')
    ax.text(df.lon.mean(),df.lat.mean(),file[10:-4],transform=ccrs.PlateCarree(),fontsize=12,ha='center',va='center',bbox=dict(facecolor='white', alpha=0.8,edgecolor='none' ))


ax.gridlines(draw_labels=True,linewidth=0.5,color='gray',alpha=0.5,linestyle='--')
```
