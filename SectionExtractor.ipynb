# Extractor de secciones

El código aquí presentado busca indicar que distintas secciones tiene cada fichero, de modo que se pueda tener en cuenta las siguientes cosas:
- Multiples secciones en un mismo fichero
- Ficheros que pertenecen a una sección pero que en el section_id no tengan los valores que le corresponden
La idea es con un bucle revisar uno a uno y dar un listado completo de todos las secciones por archivo

```python
# Importamos los paquetes necesarios
import numpy as np
import xarray as xr
import os
```

```python
# Escribimos la ruta con los datos directamente descargados
carpeta = "./Data/direct_downloads/"

# Recorremos cada carpeta y cada archivo
for carpetas in os.listdir(carpeta):
    for archivos in os.listdir(carpeta + carpetas):
        # Imprimimos nombre del archivo
        print(archivos)
        
        # Extraemos la ruta completa
        ruta = os.path.join(carpeta, carpetas, archivos)
        
        # Abrimos el dataset y extraemos los distintos valores de section_id
        ds = xr.open_dataset(ruta)
        try:
            assert "section_id" in ds.data_vars
        except AssertionError:
            print(f'{archivos} no contiene la variable section_id')
            print('----------------------------')
            continue
        
        valores = np.unique(ds.section_id)

        
        print(f'sections: {valores}')
        print('----------------------------')
```
